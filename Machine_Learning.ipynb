```python
# Menghapus baris yang mengandung nilai yang hilang
df = df.dropna()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Melakukan normalisasi pada dataset
normalized_df = scaler.fit_transform(df)
```

```python
# Mengembalikan hasil normalisasi ke dalam DataFrame
new_df = pd.DataFrame(normalized_df, columns=df.columns)
```

```python
new_df.head()
```

```python
# Memisahkan fitur (X) dan target (y)
X = new_df.drop(columns=['Potability'])
y = new_df['Potability']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
# Membagi dataset menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
from sklearn.tree import DecisionTreeClassifier
# membuat model dengan decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
# Mengevaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, X, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```
