```python
import pandas as pd
```

```python
df= pd.read_csv('/content/water_potability.csv')
```

```python
# Menghapus baris yang mengandung nilai yang hilang
df = df.dropna()
```

```python
df.info()
```

```python
df.head()
```

```python
# Memisahkan fitur (X) dan target (y)
X = df.drop(columns=['Potability'])
y = df['Potability']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
# Membagi dataset menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn import tree
# membuat model dengan decision tree classifier
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
# Mengevaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, X, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

```python
# Data untuk prediksi manual (contoh)
manual_input = [[6.79, 187.78, 15300.83,7.48, 316.95, 498.66, 19.60, 55.40, 5.16]]
```

```python
# Melakukan prediksi
prediction = clf.predict(manual_input)
print("Hasil prediksi:", prediction)
```

```python
from sklearn.tree import export_graphviz
import graphviz

# Ekspor pohon keputusan ke dalam format dot
dot_data = export_graphviz(clf, out_file=None, feature_names=X.columns,
                           class_names=['Non-Potable', 'Potable'], filled=True);

# Membuat objek grafik dari data dot
graph = graphviz.Source(dot_data);

# Menyimpan gambar pohon keputusan dalam file PDF
graph.render("decision_tree", format='pdf')
```
